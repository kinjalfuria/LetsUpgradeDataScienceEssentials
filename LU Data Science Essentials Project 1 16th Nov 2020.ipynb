```python

#importing the libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python

pubg = pd.read_csv("pubg - Dr. Darshan Ingle.csv")
pubg.head()
```

```python

dataType = pubg.dtypes
print('Data type of each column :\n')
print(dataType)
```

```python

# dataframe.describe()

numerical_cols_summary = pubg.describe()
numerical_cols_summary
```

```python

#kills -> no. of enemy players killed, teamkills -> no. of team players killed

pubg["total_kills"] = pubg["kills"] + pubg["teamKills"]
average_kills = pubg["total_kills"].mean()
print("Average kills : ",average_kills)

```

```python

kill_99_per = pubg["total_kills"].quantile(0.99)
print("99% of people killed : ",kill_99_per)
```

```python

most_kills = pubg["total_kills"].max()
print("Most Kills ever recorded : ",most_kills)
```
